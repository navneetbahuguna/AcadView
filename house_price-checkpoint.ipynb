# 3. Univariate Analysis


```python
sas.distplot(data['bedrooms'], hist = False, kde = True)
```

```python
#above graph we can conclude that max people have 2-3 bedroom
```

```python
sas.distplot(data['price'], hist=False, kde=True)

#in this graph we can see that max price is 10000000
```

```python
sas.distplot(data['floors'])
```

```python
sas.distplot(data['view'])
```

```python
data['new_price'] = (data['price']<=5000000).astype('int')
sas.distplot(data['new_price'])

# with this graph it is proof that max people buy atmost 1000000 rs
```

# 4. Bivariate Analysis

```python
sas.jointplot(x = 'new_price', y = 'sqft_living', data = data)

# we can easily find that max people purchase less than 3000 sqft
```

```python
sas.countplot(x='new_price', data=data)
```

```python
sas.jointplot(x = 'new_price', y = 'bedrooms', data = data)
```

```python
sas.boxplot(x='new_price', y='floors', data=data)

#by this graph we conclude that max people are living in maximum 2 floor
```

```python
sas.boxplot(x = 'new_price', y = 'bedrooms', data = data)
```

```python
sas.jointplot(y='new_price', x='bedrooms', data=data, kind='scatter')
# in this graph we can find that max people have less than 5 bedrooms.also we can easily find the outleirs
```

# 5. Multivariate Analysis


```python
sas.lmplot(x='bathrooms', y='bedrooms',data=data, hue='new_price')
```

```python
sas.factorplot(x='new_price', y='sqft_living', col='bedrooms', data=data, col_wrap=4, size=3)

# bedrooms 5,6 have 10000-4000 sqft living space
```
