```python
# c.) Once again , Import the dataset and this time replace the missing null values with mean of the column series. 

new = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')

print(new.isnull().sum())
print('£'*50)
print('MEAN OF ALL FEATURE')
print(new.mean())
```

```python
new.fillna(new.mean(), inplace = True)
print(new.isnull().sum())
```

```python
new.fillna(new.median(), inplace = True)
print(new.isnull().sum())
```

```python
new.fillna(new.mode(), inplace=True)
print(new.isnull().sum())
```
